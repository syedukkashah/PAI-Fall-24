```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("dm_office_sales.csv")
df.head()
```

```python
sns.scatterplot(x = 'salary', y = 'sales', data = df)

```

```python
plt.figure(figsize=(12,4))
sns.scatterplot(x='salary', y='sales', data=df)
```

```python
sns.scatterplot(x = 'salary', y ='sales', data = df, hue = 'level of education')
```

```python
sns.scatterplot(x = 'salary', y = 'sales', data = df, size = 'salary')
```

```python
sns.scatterplot(x='salary', y='sales', data=df, s=200)
```

```python
sns.scatterplot(x = 'salary', y = 'sales', alpha = 0.2, data=df)
```

```python
sns.scatterplot(x = 'salary', y = 'sales', hue = "division", style= "level of education", data = df)
```

```python
sns.rugplot(x = 'salary', data = df)
```

```python
sns.displot(df['salary'])
```

```python
sns.set_theme(style = 'whitegrid')
sns.displot(df['salary'], bins = 20, color = 'r', edgecolor = 'black', kde = True)
```

```python
sns.countplot(data=df, x="division")
```

```python
plt.figure(figsize=(10,4))
sns.countplot(data=df, x="division")
plt.ylim(90,260) #get current y-limits
```

```python
plt.figure(figsize = (10,4))
sns.countplot(data=df, x = 'level of education', hue = 'division')
```

```python
plt.figure(figsize= (10,4))
sns.barplot(data = df, x = 'level of education', y = 'salary', estimator = np.mean, errorbar = "sd")
```

```python
sns.barplot(data = df, x = 'level of education', y = 'salary', estimator = np.mean, errorbar = 'sd', hue = 'division')
```
