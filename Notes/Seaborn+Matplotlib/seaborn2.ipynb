```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = sns.load_dataset("iris")
df
```

```python
sns.histplot(data=df, x = 'sepal_length')
plt.show()
```

```python
sns.histplot(data = df, x = 'sepal_length', bins = 20, color = 'g')
plt.show()
```

```python
tips = sns.load_dataset('tips')
sns.boxplot(x='day', y = 'total_bill', data=tips)
plt.show()
```

```python
sns.scatterplot(data=tips, x='total_bill', y='size', hue = 'total_bill') #style = 'col name' will add markers to points
plt.show()
```
