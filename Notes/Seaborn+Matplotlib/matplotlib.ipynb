```python
import matplotlib.pyplot as plt
import numpy as np
```

Bar charts & pie charts

```python
heights = np.random.randint(160, 180, 20) #cm
```

```python
heights
```

```python
names = [
    "Alexander", "Olivia", "Benjamin", "Charlotte", "William",
    "Sophia", "James", "Isabella", "Michael", "Amelia",
    "Elijah", "Mia", "Ethan", "Harper", "Lucas",
    "Evelyn", "Henry", "Abigail", "Jackson", "Emily"
]
```

```python
plt.figure(figsize=(18,10))
plt.bar(names,heights,color='hotpink', edgecolor = 'blue', label = 'heights', linewidth = 2)
plt.xlabel('names')
plt.ylabel('heights')
plt.xticks(rotation=60, ha='left')
plt.legend(loc = 'upper left') #when inserting legend, always add a label to your graph/plot
plt.show()
```

```python
plt.figure(figsize=(6,8))
plt.pie(heights, labels=names)
plt.show(),
```

```python

plt.figure(figsize=(14,7)) #always define figsize first
plt.subplot(1,2,1) #subplot (1 row, 2 cols, 1st subplot)
color = ['blue','red', 'magenta']
edgecolor = ['yellow','green', 'brown']
plt.bar(names,heights, color = color, edgecolor = edgecolor,  linewidth = 1.75)
plt.title('Distribution of Heights')
plt.xlabel('Students')
plt.ylabel('Heights')
plt.xticks(rotation = 50, ha='right') #ha = horizontal alignment
plt.legend('Students', loc = 'upper left') #legend params(name of X or Y. loc = 'upper left/upper right/lower left/lower right)
plt.subplot(1,2,2)
plt.title('Pie Chart Visualization')
plt.pie(heights, labels = names, autopct="%0.01f%%")
plt.tight_layout() #adjusts padding in subplots
plt.show()
```

```python
ex = []
for i in range(0,20):
    ex.append(0)
    
ex[0] = 0.8
print(ex)
```

pie chart shows the part-to-whole relationship between variables, and is best for one string and one numeric variable. We can see the height distribution is not visualized best in a pie chart

pie charts should be used only for simple compositions — never for comparisons or distributions.


```python
plt.figure(figsize=(8,6))
colors = plt.cm.plasma(np.linspace(0,1,len(names))) #choose color map (plt.cm.plasma/inferno/viridis/coolwarm(np.linspace(0,1,len of data)))
plt.title('Pie Chart Visualization')
plt.pie(heights, labels = names, autopct="%0.01f%%", colors = colors, explode = ex, shadow = True)
plt.show()
```

```python
plt.figure(figsize = (14,7))
colors = plt.cm.inferno(np.linspace(0,1,len(names)))
plt.title('Height Distribution')
plt.xlabel('Students')
plt.ylabel('Heights')
plt.bar(names, heights, color = colors)
plt.xticks(rotation = 50, ha = 'right')
plt.tight_layout() #adjusts padding
plt.show()
```

```python
#multiple bar graphs
plt.figure(figsize = (12,10))
x = ['python', 'c++', 'c', 'java']
y = [20, 35, 44, 26]
z = [92, 48, 57, 45]
plt.bar(x,z,color='r', edgecolor = 'r', linewidth = 0.6, label = 'AI Students') #the larger y-axis values should be plotted first so they don't overlap the smaller values
plt.bar(x,y,color='b', edgecolor = 'r' , linewidth = 0.6, label = 'CS Students')
plt.title('Multiple Bar Charts')
plt.xlabel('lanugages')
plt.ylabel('users')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#side by side bar graphs
categories = ['Category A','Category B','Category C','Category D']
ds_1 = [25,40,35,35]
ds_2 = [30,45,20,45]
#set width of each bar
bar_width = 0.35
#generate array of equally spaced values for x-axis
x = np.arange(len(categories))
plt.bar(x-bar_width/2, ds_1, bar_width, label='Dataset', color = 'b')
plt.bar(x+bar_width/2, ds_2, bar_width, label = "Dataset 2", color = 'g')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Side by Side Bar graph')
plt.xticks(x, categories) #set tick labels to category names
plt.legend()
plt.show()
```

```python
#horizontal bar graph
cities = [
    "Tokyo", "Delhi", "Shanghai", "São Paulo", "Mumbai",
    "Mexico City", "Beijing", "Cairo", "Dhaka", "Osaka",
    "New York City", "Karachi", "Buenos Aires", "Chongqing",
    "Istanbul", "Kolkata", "Manila", "Lagos", "Rio de Janeiro", "Los Angeles"
]
populations = [
    37.4, 28.51, 25.58, 21.65, 21.52,
    21.52, 20.04, 20.0, 19.58, 19.28,
    18.82, 15.4, 14.97, 14.83,
    14.8, 14.69, 13.92, 13.78, 13.46, 12.82
]
plt.figure(figsize=(14,8))
plt.barh(cities, populations,color = 'hotpink', label = 'City Population', linewidth = 2, edgecolor = 'b')
plt.title('Population of Cities')
plt.xlabel('Population')
plt.ylabel('Cities')
plt.legend(title = 'legend', loc = 'upper right')
plt.show()
```

```python
flavors = [
    "Vanilla", "Chocolate", "Strawberry", "Mint Chocolate Chip", "Cookie Dough"
]
scoops_sold = [500, 450, 300, 275, 320]
c = plt.cm.viridis(np.linspace(0,1,len(flavors)))
plt.figure(figsize = (40,8))
plt.pie(scoops_sold, labels=flavors, colors=c, autopct = '%1.1f%%',shadow = True, startangle = 100)
plt.legend(title = 'Key: ', loc = 'upper left', fontsize = 'small', ncol = 2) #ncol = num of cols legend is displayed in
plt.tight_layout()
plt.show()
```

```python
freq = [10, 20, 18, 30]
age_ranges = ['18-25', '26-30', '31-35', '36+']
plt.figure(figsize = (10,6))
plt.title('Age Frequency: ')
plt.pie(freq, labels = age_ranges, autopct = '%1.1f%%', shadow = True)
plt.legend(fontsize = 'small', ncol = 2)
plt.show()
```

```python

bar_width = 0.2
plt.figure(figsize = (10,10))
plt.bar(x-bar_width / 2, ds_1, width = bar_width, color = 'y', label = 'dataset 1')
plt.bar(x+bar_width/2, ds_2, width = bar_width, color = 'm', label = 'dataset 2')
plt.title('side by side bar graph')
plt.xlabel('categories')
plt.xticks(x, categories)
plt.ylabel('heights')
plt.legend(fontsize = 'large', loc = 'upper left', ncol = 2, title = 'KEY:')
plt.tight_layout()
plt.show()
```

```python
x = np.arange(7)
straight_line = x
parabola = x**2
cubic = x**3
plt.subplot(2,1,1)
plt.plot(x,straight_line)
plt.title('Straight Line')
plt.subplot(2,2,3)
plt.plot(x,parabola)
plt.title('Parabola')
plt.subplot(2,2,4)
plt.plot(x,cubic)
plt.title('cubic')
plt.tight_layout()
plt.show()
```

```python
#multiple line graphs
overs = range(1,11)
Pakistan = [5,15,18,30,50,55,59,65,65,74]
Aus = [10,25,36,36,45,55,59,65,80,100]
plt.plot(overs, Pakistan, color = 'g', label = 'Pakistan')
plt.plot(overs,Aus, color = 'y', label = 'Australia')
plt.legend(title = 'Teams', loc = 'upper left')
plt.xlabel('Overs')
plt.ylabel('Runs')
plt.show()
```

```python
overs = range(1,11)
Pak = [5,15,18,30,50,55,59,65,65,74]
NZ = [10,25,36,36,45,55,59,65,80,100]
plt.plot(overs,Pak, linestyle='--', color = 'g', marker = 'o', label = 'Pakistan')
plt.plot(overs, NZ, linestyle = ':', marker = 'x', color = 'b', label = 'New Zealand')
plt.legend(title = 'Teams', loc = 'upper left')
plt.xlabel('Overs')
plt.ylabel('Runs')
plt.show()
```

Scatterplot

```python
days = range(1,8)
expenses = [250,500,100,10,300,360,575]
plt.scatter(days,expenses)
sizes = [400,200,500,100,50,250,350]
colors = ['r','b','y','g','black','pink','purple']
plt.scatter(days,expenses, s = sizes, c = colors)
plt.xlabel('Days')
plt.ylabel('Expenses')
plt.title('Expense Report', fontsize = 15)
plt.show()
```

```python
#changing markers
days = range(1,8)
expenses = [250,500,100,10,300,360,575]
plt.scatter(days,expenses)
sizes = [400,200,500,100,50,250,350]
colors = ['r','b','y','g','black','pink','purple']
plt.scatter(days,expenses, s = sizes, c = colors, marker = '*', edgecolor = 'g')
plt.xlabel('Days')
plt.ylabel('Expenses')
plt.title('Expense Report', fontsize = 15)
plt.show()
```

Histograms -> useful for visualizing frequency distributions 

```python
freq = [1, 1, 1, 2, 2, 2, 2, 2, 3, 3, 4, 4, 4, 4, 5]
plt.hist(freq, width = 0.5, color = 'r', edgecolor = 'black', linewidth = 2)
plt.title('Frequency of nums')
plt.xlabel('nums')
plt.ylabel('occurences')
plt.show()
```

```python
freq = [1, 1, 1, 2, 2, 2, 2, 2, 3, 3, 4, 4, 4, 4, 5]
plt.hist(freq, "auto", (1,8),width = 0.5, color = 'r', edgecolor = 'black', linewidth = 2)
plt.title('Frequency of nums')
plt.xlabel('nums')
plt.ylabel('occurences')
plt.show()
```

```python
marks = [np.random.randint(1, 100) for marks in range(40)]
```

```python
print(marks)
```

```python
#bins reduce bars in histogram based on range
#cumulative = 1 (displayed in ascending order)
plt.hist(marks, bins=10, edgecolor = 'y', color = 'g', cumulative=1)
plt.title('Histogram of Marks')
plt.xlabel('marks')
plt.ylabel('num of students')
plt.show()
```

```python
plt.hist(marks, bins=10, edgecolor = 'y', color = 'g', cumulative= -1)
plt.title('Histogram of Marks')
plt.xlabel('marks')
plt.ylabel('num of students')
plt.show()
```

```python
plt.hist(marks, bins=10, edgecolor = 'b', color = 'g', cumulative= -1, histtype = 'step')
plt.title('Histogram of Marks')
plt.xlabel('marks')
plt.ylabel('num of students')
plt.show()
```

```python
plt.hist(marks, bins=10, edgecolor = 'b', color = 'g', cumulative= -1, histtype = 'stepfilled')
plt.title('Histogram of Marks')
plt.xlabel('marks')
plt.ylabel('num of students')
plt.show()
```

```python
#changing orientation
plt.hist(marks, bins=10, edgecolor = 'b', color = 'g', cumulative= -1, histtype = 'step', orientation='horizontal')
plt.title('Histogram of Marks')
plt.xlabel('marks')
plt.ylabel('num of students')
plt.show()
```
