```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
iris = pd.read_csv('Iris.csv')
```

```python
inputs = iris.drop('Species', axis = 1)
inputs.drop('Id', axis=1, inplace=True)
target = iris['Species']
```

```python
inputs.head()
```

```python
target.head()
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(inputs, target)
```

```python
inputs_train, inputs_test, target_train, target_test = train_test_split(inputs, target, train_size=0.2, random_state=0)
```

```python
predictions = knn.predict(inputs_test)
```

```python
accuracy_score(target_test, predictions)
```
