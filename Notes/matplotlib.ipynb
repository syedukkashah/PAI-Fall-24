```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

x = ["python", "c", "c++", "java"]
y = [85,70,60,82]
plt.bar(x,y)
plt.show()
```

```python
plt.xlabel("languages")
plt.ylabel("No")
plt.title("FAST NUCES")
plt.bar(x,y)
plt.show()
```

```python
#width adjustment
plt.xlabel("languages", fontsize = 25) #we can adjust font size in x & y labels 
plt.bar(x,y,width=0.4)
plt.show()
```

```python
plt.bar(x,y,width=0.4,color="y") #color changed
plt.show()
```

```python
#multiple colors can be init to bars by passing a list to the bar function parameters
c = ["y", "r", "m", "b"]
plt.bar(x,y,width=0.4,color = c, edgecolor = "g", linewidth = 1, label = "popularity") #borders can be added around bars, legend is added by label
plt.legend()
plt.show()
```

```python
#multiple bar graphs
z = [20,30,40,50]
plt.xlabel("languages", fontsize = 10)
plt.ylabel("No")
plt.bar(x,y,width = 0.4, label = "popularity", linewidth = 1, color = "r", edgecolor = "g")
plt.bar(x,z, width = 0.4, linewidth = 1, edgecolor = "g", label = "popularity2", color = "b")
plt.legend()
plt.show()
```

```python
#scatter plot
x = np.arange(1,8)
y = [2,3,1,4,5,3,6]
colors = ['r','y','g','b','r','y','b']
sizes = [400,200,300,150,500,350,200]
plt.scatter(x,y, c = colors, s = sizes, marker = "*", edgecolor = 'g') #we can add markers to illustrate the points
plt.xlabel("Day")
plt.ylabel("No")
plt.title("Scatter plot", fontsize = 15)
plt.show()
```

```python
#histogram
no = np.random.randint(20,61,50)
plt.hist(no, color = "b", edgecolor = "r")
plt.xlabel('student marks')
plt.ylabel('frequency')
plt.title('PAI class final grades')
plt.show()
```

```python
l = [20,25,30,35,40,45,50,55,60]
plt.hist(no,bins = l,edgecolor = 'r', color = 'b') #we ensure that bar graphs are illustrated clearly with bins 
```

```python
plt.hist(no, bins = "auto", range = (0,80), edgecolor = 'r', color = 'b', cumulative = -1) #descending order
plt.show()
```

```python
plt.hist(no, bins = "auto", range = (0,80), edgecolor = 'r', color = 'b', cumulative = 1) #ascending order
plt.show()
```

```python
plt.hist(no, bins = l, edgecolor = 'r', color = 'b', bottom = 15) #bottom(origin) both axes will be starting from 15
plt.show()
```

```python
plt.hist(no, bins = l, edgecolor = 'r', histtype="step", color = 'b', label = "yeh label hai")
plt.xlabel("x label")
plt.ylabel("y ka label")
plt.legend()
plt.show()
```

```python
#histtype = "stepfilled" will fill the bar with mentioned color
plt.hist(no, bins = l, edgecolor = 'r', histtype="stepfilled", color = 'b', label = "yeh label hai")
plt.show()
```

```python
#orientation can be set
plt.hist(no, bins = l, edgecolor = 'r', color = 'b', label = "yeh label hai", orientation="horizontal")
plt.xlabel("x label")
plt.ylabel("y ka label")
plt.show()
```

```python
#Pie chart
x = [10,20,30,40]
y = ['c', 'c++', 'java', 'python']
plt.pie(x)
plt.show()
```

```python
ex = [0.05, 0.05, 0.05, 0.05]
plt.pie(x, labels = y, explode=ex)
plt.show()
```

```python
ex = [0.4,0,0,0]
color = ['r','g','b','y']
plt.pie(x,labels = y, explode = ex, colors= color)
plt.show()
```

```python
plt.pie(x,labels = y, explode = ex, colors = color, autopct = "%0.01f%%")
plt.show()
```

```python
#Stackplot
x = [1,2,3,4,5]
area = x
plt.stackplot(x,area,color = 'm')
plt.xlabel("yeh x hai")
plt.ylabel("yeh y hai")
plt.show()
```

```python
#illustrate multiple areas in a stackplot
x = [1,2,3,4,5]
a1 = [2,3,2,5,4]
a2 = [2,3,4,5,6]
a3 = [1,3,2,4,2]
c = ['m', 'b', 'r']
l = ['a1', 'a2', 'a3']
plt.stackplot(x,a1,a2,a3, colors = c, labels = l)
plt.legend()
plt.show()
```

```python
x = [10,20,40,50,60,70]
plt.boxplot(x)
plt.xlabel("wat da hell is this")
plt.show()
```

```python
plt.boxplot(x,vert = False)
plt.show()
```

```python
plt.boxplot(x,widths = 0.5)
plt.show()
```
