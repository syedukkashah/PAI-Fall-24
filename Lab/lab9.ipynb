```python
import matplotlib.pyplot as plt
import numpy as np
names = [
    "Alice", "Bob", "Charlie", "David", "Eva",
    "Fiona", "George", "Hannah", "Isaac", "Jasmine",
    "Kevin", "Lily", "Mason", "Nina", "Oliver",
    "Penny", "Quinn", "Ryan", "Sophia", "Thomas"
]

heights = [
    150, 162, 175, 180, 155,
    167, 170, 158, 165, 172,
    177, 185, 160, 168, 173,
    182, 159, 176, 169, 174
]

plt.figure(figsize=(14, 7))
plt.subplot(1,2,1)
colors = plt.cm.viridis(np.linspace(0, 1, len(names)))
plt.bar(names, heights, color=colors)
plt.xticks(rotation=45, ha='right')
plt.xlabel('Students')
plt.ylabel('Height (cm)')
plt.title('Heights of 20 Students')
plt.subplot(1, 2, 2) 
plt.pie(heights, labels=names, autopct='%1.1f%%', startangle=140, colors=colors)
plt.title('Height Distribution of Students')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
flavors = [
    "Vanilla",
    "Chocolate",
    "Strawberry",
    "Mint Chocolate Chip",
    "Cookies and Cream",
    "Rocky Road",
    "Pistachio",
    "Mango",
    "Butter Pecan",
    "Cookie Dough"
]
scoops_sold = [
    150,
    120,
    80,
    60,
    90,
    75,
    50,
    30,
    45,
    55
]
plt.figure(figsize=(10, 8))
plt.pie(scoops_sold, labels=flavors, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Ice Cream Sales by Flavor')
plt.axis('equal')
plt.show()
```

```python

math_marks = [78, 88, 94, 65, 70, 85, 90, 72, 68, 80]
science_marks = [82, 85, 89, 75, 74, 88, 92, 70, 69, 78]
plt.figure(figsize=(8, 6))
plt.scatter(math_marks, science_marks, color='blue', label='Marks Comparison', s=100)
plt.title("Scatter Plot: Mathematics vs Science Marks", fontsize=16)
plt.xlabel("Mathematics Marks", fontsize=14)
plt.ylabel("Science Marks", fontsize=14)
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
